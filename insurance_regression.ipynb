```python
from sklearn.datasets import load_diabetes
import seaborn as sns
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

### The insurance.csv file imported into the notebook

```python
file_path = 'insurance.csv'
data = pd.read_csv(file_path)
insurance_data = pd.DataFrame(data)
insurance_data.head()
```

### Relevant columns to determine how age affects insurance costs.

```python
age = insurance_data.iloc[:,:1].values
charges = insurance_data.iloc[:,-1].values
```

### A scatter plot with age on x-axis and charges on the y-axis

```python
plt.figure()
plt.scatter(age, charges, color = 'orange')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age verses Insurance Cost')
plt.show()
plt.close()
```

### A linear Regression model fit and make predictions on the data

```python

insurance_model = LinearRegression()
insurance_model.fit(age, charges)
```

### Another scatter plot with the best-fit line

```python
predicted_charges = insurance_model.predict(age)

plt.figure()
plt.scatter(age, charges, color = 'orange')
plt.plot(age, predicted_charges, color='b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age verses Insurance Cost')
plt.show()
plt.close()
```

### Another scatter plot with the best-fit line

```python
assumed_charges = insurance_model.predict(age)

plt.figure()
plt.scatter(age, charges, color = 'orange')
plt.plot(age, assumed_charges, color='b')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age verses Insurance Cost')
plt.show()
plt.close()
```

### Predict an unknown value

```python

unk_x = [[100]] 

x_pred = np.append(age, unk_x).reshape(-1,1)
y_pred = insurance_model.predict(x_pred)

plt.scatter(age,charges,color = 'b')
plt.plot(x_pred,y_pred,color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

### Discussion

The linear regression model of age versus charges suggest that age could affect insurance cahrges but not the sole factor that determines the increase in insurance cost. A person's lifestyle seem like other contributing factors.
